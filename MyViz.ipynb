```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'file_path' with the actual path to your CSV file
file_path = '../data/df_messi.csv'

# Define the columns you want to read from the CSV
columns_to_use = [
    'player.name', 'location_x', 'location_y', 'shot.outcome.name', 'shot.outcome.id'
]

# Read data from the CSV file using specified columns
data = pd.read_csv(file_path, usecols=columns_to_use)

# Filter shots that are on target (on goal)
on_target_shots = data[data['shot.outcome.id'] == 97]  # Assuming 97 represents shots on target

# Create a football field plot
fig, ax = plt.subplots(figsize=(10, 7))

# Plotting football field lines (same as previous code)

# Plotting heatmap of on-target shots
hb = ax.hexbin(
    on_target_shots['location_x'],  # X-axis data (field width)
    on_target_shots['location_y'],  # Y-axis data (field length)
    gridsize=20,                    # Adjust grid size as needed
    cmap='YlOrRd',                  # Choose a colormap for heatmap
    mincnt=1                        # Minimum count to display a hexagon
)

# Add colorbar for the heatmap
cb = fig.colorbar(hb, ax=ax)
cb.set_label('Shots Density')

# Set plot title and labels
plt.title('Heatmap of On-Target Shot Locations on Football Field')
plt.xlabel('X-axis (Width of the Field)')
plt.ylabel('Y-axis (Length of the Field)')
plt.xlim(0, 120)  # Field width (adjust according to your data)
plt.ylim(0, 80)   # Field length (adjust according to your data)
plt.gca().invert_yaxis()  # Invert y-axis to match football field perspective
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Replace 'file_path' with the actual path to your CSV file
file_path = '../data/df_messi.csv'

# Define the columns you want to read from the CSV
columns_to_use = [
    'player.name', 'location_x', 'location_y', 'shot.outcome.name', 'shot.outcome.id'
]

# Read data from the CSV file using specified columns
data = pd.read_csv(file_path, usecols=columns_to_use)

# Filter shots that are on target (on goal)
on_target_shots = data[data['shot.outcome.id'] == 97]  # Assuming 97 represents shots on target

# Create a soccer field plot
fig, ax = plt.subplots(figsize=(10, 7))

# Plotting soccer field lines
plt.plot([0, 0, 120, 120, 0], [0, 80, 80, 0, 0], color='black')  # Outer boundary
plt.plot([60, 60], [0, 80], color='black')  # Center line

# Create a heatmap of on-target shots with increased bins for finer resolution
heatmap, xedges, yedges = np.histogram2d(
    on_target_shots['location_x'],
    on_target_shots['location_y'],
    bins=[np.arange(0, 121, 5), np.arange(0, 81, 5)]  # Adjust bin sizes for finer resolution
)

extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
plt.imshow(heatmap.T, extent=extent, origin='lower', cmap='YlOrRd', alpha=0.7)

# Set plot title and labels
plt.title('Heatmap of On-Target Shot Locations on Soccer Field (Finer Resolution)')
plt.xlabel('Width of the Field')
plt.ylabel('Length of the Field')
plt.xlim(0, 120)
plt.ylim(0, 80)
plt.gca().invert_yaxis()  # Invert y-axis to match football field perspective
plt.grid(False)
plt.show()
```
