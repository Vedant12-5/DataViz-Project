```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'file_path' with the actual path to your CSV file
file_path = '../data/df_messi.csv'

# Define the columns you want to read from the CSV
columns_to_use = [
    'player.name', 'location_x', 'location_y', 'shot.outcome.name'
]

# Read data from the CSV file using specified columns
messi_data = pd.read_csv(file_path, usecols=columns_to_use)

# Plotting shot locations
plt.figure(figsize=(8, 6))
plt.scatter(messi_data['location_x'], messi_data['location_y'], color='red')
plt.title('Shot Locations by Messi')
plt.xlabel('X-axis (Width of the Field)')
plt.ylabel('Y-axis (Length of the Field)')
plt.xlim(0, 120)  # Field width (adjust according to your data)
plt.ylim(0, 80)   # Field length (adjust according to your data)
plt.gca().invert_yaxis()  # Invert y-axis to match football field perspective
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

# Plotting football field lines
ax.plot([0, 0, 120, 120, 0], [0, 80, 80, 0, 0], color='black')  # Outline of the field
ax.plot([60, 60], [0, 80], color='black')  # Halfway line
ax.scatter(100, 40, color='black')  # Penalty spot (right)
ax.scatter(20, 40, color='black')   # Penalty spot (left)
ax.scatter(60, 40, color='black')   # Center spot
circle = plt.Circle((60, 40), 9.15, color='black', fill=False)  # Center circle
ax.add_patch(circle)
```

```python
for index, shot in messi_data.iterrows():
    x = shot['location_x']
    y = shot['location_y']
    outcome = shot['shot.outcome.name']
    color = 'red' if outcome == 'Goal' else 'blue'  # Color based on shot outcome
    ax.scatter(x, y, color=color)

# Set plot title and labels
plt.title('Shot Locations on Football Field')
plt.xlabel('X-axis (Width of the Field)')
plt.ylabel('Y-axis (Length of the Field)')
plt.xlim(0, 120)  # Field width (adjust according to your data)
plt.ylim(0, 80)   # Field length (adjust according to your data)
plt.gca().invert_yaxis()  # Invert y-axis to match football field perspective
plt.grid(False)
plt.show()
```
